```python
# Installation de lxml (dans un notebook Jupyter uniquement)
!pip install -q requests beautifulsoup4 lxml


# Import des bibliothèques nécessaires
import requests
from bs4 import BeautifulSoup

# URL de la page Wikipedia de la Ligue 1 2019-2020
url_ligue_1 = "https://fr.wikipedia.org/wiki/Championnat_de_France_de_football_2019-2020"

# Faire une requête GET pour récupérer le contenu de la page
response = requests.get(url_ligue_1)

# Vérifier que la requête a réussi (statut 200)
if response.status_code == 200:
    print("Requête réussie !")
else:
    print(f"Erreur lors de la requête : {response.status_code}")

# Obtenir le contenu brut de la page
code_source = response.content

# Analyse du code source avec BeautifulSoup en utilisant le parser (analyseur) 'lxml'
page = BeautifulSoup(code_source, "lxml")

# Affichage d'un extrait de la page pour vérifier
print(page.title.text)

print(page.find("title")) # renvoie que la première occurence sinon faire : findAll
print(page.findAll("table")) # renvoie une liste de tous les tableaux 
print(page.table.text) # renvoie en str le contenu du premier tableau
```
